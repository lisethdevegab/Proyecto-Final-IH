```python
def load_data(file):
    data = pd.read_csv(file, encoding='utf8')
    df = pd.DataFrame(data)
    return df
```

```python
acnur = load_data('poblacion_acnur.csv')
```

```python
acnur.shape
```

```python
acnur.columns
```

```python
#Daba error por el caracter especial al momento de elegir columna, se reemplaza con diccionario

acnur['Value'].replace({'*': '0'}, inplace=True)
```

```python
acnur.info()
```

```python
acnur['Value'] = acnur['Value'].astype(int)
```

```python
acnur.info()
```

```python
#poblacion = acnur.groupby(by=['Origin', 'Year', 'Population type']).agg({'Value': [sum]})

origen = acnur.pivot_table(index=["Origin", "Population type"], columns=["Year"], values=["Value"], aggfunc=np.sum, fill_value=0)

origen.head()
```

```python
origen.columns = origen.columns.map('{0[0]}_{0[1]}'.format)

origen.head()
```

```python
origen = origen.reset_index()

origen.head(10)
```

```python
origen.columns
```

```python
origen = origen.rename(columns={'Value_2016': 2016, 'Value_2017': 2017, 'Value_2018': 2018})

origen.head()
```

```python
origen.to_csv('origen.csv', index=False)
```
