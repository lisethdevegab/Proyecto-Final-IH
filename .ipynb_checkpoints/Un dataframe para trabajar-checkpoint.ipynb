# 2) Procedemos a la limpieza de los datos

```python
#Esta función es especial para el DataFrame para proceder a la separación de datos
def separate_col(df, column):
    df[['Longitude','Latitude']] = df.pop('Location Coordinates').str.split(',\s*', n=1, expand=True)
    return df
```

```python
separate_col(df2,['Location Coordinates'])

df2.head()
```

```python
#Los datos nulos, los sustituimos por la mediana de cada columna que nos interesa y la convertimos en int
def medianfornul(df, column):
    median_value= df[column].median()
    df[column]= df[column].fillna(median_value)
    df[column]= df[column].astype(int)
    return df

```

```python
medianfornul(df2,['Number of Children'])
```

```python
medianfornul(df2,['Minimum Estimated Number of Missing'])
```

```python
medianfornul(df2,['Number Dead'])
```

```python
#Remover espacios en blancos, puntos
def remove_nonecessary(df):
    whiteandspecial = lambda x: x.strip(r'.!?" " \n\t') if isinstance(x, str) else x
    return df.applymap(whiteandspecial)
```

```python
remove_nonecessary(df2)
```

```python
#Reemplazar valores de una columna pasando diccionario

def new_values(df, column, dictionary):
    df[column].replace(dictionary, inplace=True)
    print(df)
```

```python
#Nuestro dato más importante que es el Total de Muertes y Desparecidos es necesario pasarlo a int

def for_int(df, column):
    df[column]= df[column].astype(int)
    return df

for_int(df2,'Total Dead and Missing')

df2.dtypes
```

```python
#Las columnas de string, sus nulos lo sustituimos por columnas correlativas

def fillna_col(df,column1, column2):
    df[column1].fillna(df[column2], inplace=True)
    return df


fillna_col(df2,'UNSD Geographical Grouping', 'Region of Incident')
```

```python
fillna_col(df2,'Migration Route', 'UNSD Geographical Grouping')
```

```python
fillna_col(df2,'Information Source', 'Source Quality')
```

```python
#En nuestros calculos, sólo Causa de Muerte contiene nulos, los borramos

def delete_null(df, column):
    df.dropna(subset=[column], inplace=True)
    return df

delete_null(df2, 'Cause of Death')
```

```python
#Calculamos los datos nulos

percent = (df2.isnull().sum()/df2.isnull().count()).sort_values(ascending=False)

percent
```

```python
#Al tener nuestra la columna de mes y año, tomamos el día.

def take_day(df, column):
    df['Reported day'] = df[column].apply(lambda date: date.split(" ")[1])
    df['Reported day'] = df['Reported day'].str.replace(",","").astype(int)
    return df

take_day(df2,'Reported Date')
```

```python
#Unimos nuestro datos de día, mes y año para hacer nuestra columna de datatime


def completadate(df, column1, column2, column3):
    df['Reported case'] = (pd.to_datetime(df[column1].astype(str) + '-' +
                                  df[column2].astype(str) + '-' +
                                  df[column3].astype(str)))
    

completadate(df2,'Reported day', 'Reported Month', 'Reported Year')

df2.head()
```
